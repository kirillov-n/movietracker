Сначала заполним пропуске в столбце, который был выбран из оценочных как оптимальный

так как во всех оценочных столбцах столбцах в этих строках пусто, заполним пропуски средней оценкой

```python
data[data['ratingImdb'].isnull()]
```

```python
data['ratingImdb'] = data['ratingImdb'].fillna(data['ratingImdb'].mean())
```

Теперь сделаем универсальный столбец с названием, в котором не будет пропусков

```python
data['name'] = data['nameRu'].fillna(data['nameOriginal'])
```

проверим на одном из фильмов

```python
data[data['imdbId'] == 'tt0019303']
```

выяснилось, что фильм по какой-то причине дублируется. нужно найти все дубликаты и устранить

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.head()
```

Теперь перейдём к удалению ненужных столбцов

```python
data = data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'nameEn', 'nameRu', 'nameOriginal', 'reviewsCount', 'ratingGoodReview',
       'ratingGoodReviewVoteCount', 'ratingKinopoisk',
       'ratingKinopoiskVoteCount', 'ratingImdbVoteCount',
       'ratingFilmCritics', 'ratingFilmCriticsVoteCount', 'ratingAwait',
       'ratingAwaitCount', 'ratingRfCritics', 'ratingRfCriticsVoteCount', 'hasImax', 'has3D', 'message'])
```
