# Парсим кинопоиск (ну типа) туториал для чайников и кастрюль

```python
import requests
import json
from tqdm import tqdm
import time
import pandas as pd
```

свой апи ключ можно получить здесь https://kinopoiskapiunofficial.tech/profile (сначала надо зарегистрироваться) и подставить в ячейку ниже между символами ' и ' (в общем записать как строку в переменную my_api_key). мой ключ на всякий случай убрала, он мне ещё пригодится

```python
my_api_key = ''
```

дальше просто запускаем

```python
headers = {
    'accept': 'application/json',
    'X-API-KEY': my_api_key,
}
```

```python
movies_list = []
```

в range(вот сюда) в ячейке ниже записываем диапазон интересующих нас айдишников

```python
for i in tqdm(range(300, 750)):
    st = str(i)
    response = requests.get(f'https://kinopoiskapiunofficial.tech/api/v2.2/films/{st}', headers=headers)
    movies_list.append(json.loads(response.text))
    time.sleep(20)
```

получаем списочек из словарей, посмотрим на него, какой же он красивый

```python
movies_list
```

сделаем из него датафрейм

```python
movies_df = pd.DataFrame(movies_list)
```

возможно как я превысили доступное количество запросов, посмотрим так ли это (если запускать сразу нормально, скорее всего такого не будет)

```python
movies_df
```

если превышение случилось, удаляем такие строчки (смотрим по столбцу кинопоиск айди)

```python
movies_cleaned = movies_df.dropna(axis='index', how='any', subset=['kinopoiskId'])
```

смотрим сколько фильмов осталось в почищенном датафрейме, записываем, какой диапазон айдишников в итоге удалось спарсить

```python
movies_cleaned
```

в названии файла указываем диапазон айдишников фильмов, которые удалось заполучить (чтобы потом было удобнее ориентироваться)

```python
movies_cleaned.to_csv('parsed_300_311.csv')  
```
