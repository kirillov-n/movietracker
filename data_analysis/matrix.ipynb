```python
import pandas as pd
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('prepared_data.csv')
```

```python
gnrs = pd.Series(data['genres'])
data['мелодрама'] = gnrs.str.contains('мелодрама', regex=False).astype(int)
data['приключения'] = gnrs.str.contains('приключения', regex=False).astype(int)
data['комедия'] = gnrs.str.contains('комедия', regex=False).astype(int)
data['музыка'] = gnrs.str.contains('музыка', regex=False).astype(int)
data['семейный'] = gnrs.str.contains('семейный', regex=False).astype(int)
data['фантастика'] = gnrs.str.contains('фантастика', regex=False).astype(int)
data['боевик'] = gnrs.str.contains('боевик', regex=False).astype(int)
data['драма'] = gnrs.str.contains('драма', regex=False).astype(int)
data['детектив'] = gnrs.str.contains('детектив', regex=False).astype(int)
data['триллер'] = gnrs.str.contains('триллер', regex=False).astype(int)
data['криминал'] = gnrs.str.contains('криминал', regex=False).astype(int)
data['фентези'] = gnrs.str.contains('фентези', regex=False).astype(int)
data['спорт'] = gnrs.str.contains('спорт', regex=False).astype(int)
data['биография'] = gnrs.str.contains('биография', regex=False).astype(int)
data['военный'] = gnrs.str.contains('военный', regex=False).astype(int)
data['история'] = gnrs.str.contains('история', regex=False).astype(int)
data['ужасы'] = gnrs.str.contains('ужасы', regex=False).astype(int)
data['вестерн'] = gnrs.str.contains('вестерн', regex=False).astype(int)
data['фильм-нуар'] = gnrs.str.contains('фильм-нуар', regex=False).astype(int)
data['мюзикл'] = gnrs.str.contains('мюзикл', regex=False).astype(int)
data['мультфильм'] = gnrs.str.contains('мультфильм', regex=False).astype(int)
data['документальный'] = gnrs.str.contains('документальный', regex=False).astype(int)
data['короткометражка'] = gnrs.str.contains('короткометражка', regex=False).astype(int)
```

```python
df = pd.crosstab(data.name, data.name)
df = df.replace(1, 0)
df['name'] = df.columns
for index, row in data.iterrows():
    common_list = []
    name = row['name']
    if row['мелодрама'] == 1: 
        same = data[data['мелодрама'] == 1]
        common_list.append(same['name'].tolist())
    if row['приключения'] == 1: 
        same = data[data['приключения'] == 1]
        common_list.append(same['name'].tolist()) 
    if row['музыка'] == 1: 
        same = data[data['музыка'] == 1]
        common_list.append(same['name'].tolist())   
    if row['семейный'] == 1: 
        same = data[data['семейный'] == 1]
        common_list.append(same['name'].tolist())   
    if row['фантастика'] == 1: 
        same = data[data['фантастика'] == 1]
        common_list.append(same['name'].tolist())   
    if row['боевик'] == 1: 
        same = data[data['боевик'] == 1]
        common_list.append(same['name'].tolist())
    if row['драма'] == 1: 
        same = data[data['драма'] == 1]
        common_list.append(same['name'].tolist())   
    if row['детектив'] == 1: 
        same = data[data['детектив'] == 1]
        common_list.append(same['name'].tolist())   
    if row['триллер'] == 1: 
        same = data[data['триллер'] == 1]
        common_list.append(same['name'].tolist())   
    if row['криминал'] == 1: 
        same = data[data['криминал'] == 1]
        common_list.append(same['name'].tolist())   
    if row['фентези'] == 1: 
        same = data[data['фентези'] == 1]
        common_list.append(same['name'].tolist())
    if row['спорт'] == 1: 
        same = data[data['спорт'] == 1]
        common_list.append(same['name'].tolist())
    if row['биография'] == 1: 
        same = data[data['биография'] == 1]
        common_list.append(same['name'].tolist())
    if row['военный'] == 1: 
        same = data[data['военный'] == 1]
        common_list.append(same['name'].tolist())
    if row['история'] == 1: 
        same = data[data['история'] == 1]
        common_list.append(same['name'].tolist())
    if row['ужасы'] == 1: 
        same = data[data['ужасы'] == 1]
        common_list.append(same['name'].tolist())
    if row['вестерн'] == 1: 
        same = data[data['вестерн'] == 1]
        common_list.append(same['name'].tolist())
    if row['фильм-нуар'] == 1: 
        same = data[data['фильм-нуар'] == 1]
        common_list.append(same['name'].tolist())
    if row['мюзикл'] == 1: 
        same = data[data['мюзикл'] == 1]
        common_list.append(same['name'].tolist())
    if row['мультфильм'] == 1: 
        same = data[data['мультфильм'] == 1]
        common_list.append(same['name'].tolist())
    if row['документальный'] == 1: 
        same = data[data['документальный'] == 1]
        common_list.append(same['name'].tolist())
    if row['короткометражка'] == 1: 
        same = data[data['короткометражка'] == 1]
        common_list.append(same['name'].tolist())
    common_list = sum(common_list, [])
    common_dict = Counter(common_list)
    for film in common_dict.keys():
        df[f"{name}"][f"{film}"] += common_dict[film]
```
